## Data Pre-Processing

```python
Unwantedcol = Flightsdf[['Airline_dot', 'Cancellation_code', 'Delay_due_carrier', 'Delay_due_weather', 'Delay_due_nas', 'Delay_due_security',
                         'Delay_due_late_aircraft']]
```

```python
Flightsdf=Flightsdf.drop(Unwantedcol,axis=1)
```

```python
100*Flightsdf.isnull().sum()/len(Flightsdf)
```

```python
def missing(Flightsdf):
    missing_nan=100*Flightsdf.isnull().sum()/len(Flightsdf)
    missing_nan=missing_nan[missing_nan>0].sort_values()
    return missing_nan
```

```python
missing_nan=missing(Flightsdf)
```

```python
sns.barplot(x=missing_nan.index,y=missing_nan)
plt.xticks(rotation=90)
plt.show()
```

```python
# Fill null values in rows where 'Cancelled' or 'Diverted' is equal to 1 with 0
Flightsdf.loc[(Flightsdf['Cancelled'] == 1) | (Flightsdf['Diverted'] == 1), :] = Flightsdf[(Flightsdf['Cancelled'] == 1) | (Flightsdf['Diverted'] == 1)].fillna(value=0)
```

```python
#dropping the less missing value rows from columns
Flightsdf=Flightsdf.dropna()
```

```python
Flightsdf.shape
```

```python
Flightsdf.isnull().sum()
```

```python
Flightsdf['Diverted'].value_counts()
```

```python
Flightsdf['Fl_date']=Flightsdf['Fl_date'].astype('datetime64')
```

```python
# Select only numerical columns for outlier detection
numerical_columns = Flightsdf.select_dtypes(include=['number'])

# Remove 'Cancelled' and 'Diverted' from numerical_columns to keep our cancelled and diverted flights for analysis
numerical_columns = numerical_columns.drop(['Cancelled', 'Diverted'], axis=1)

# Calculating the first and third quartiles
Q1 = numerical_columns.quantile(0.25)
Q3 = numerical_columns.quantile(0.75)

# Calculate the IQR (Interquartile Range)
IQR = Q3 - Q1

# Defining the threshold for outliers
threshold = 1.5

# Identifying the outliers
outliers = ((numerical_columns < (Q1 - threshold * IQR)) | (numerical_columns > (Q3 + threshold * IQR))).any(axis=1)

# Removing outliers
Flightsdf = Flightsdf[~outliers]
```

```python
# Extract year and month from the 'FL_DATE' column
Flightsdf['Year'] = Flightsdf['Fl_date'].dt.year
Flightsdf['Month'] = Flightsdf['Fl_date'].dt.month

# Pivot table to aggregate flight volume by year and month
flightvol_months = Flightsdf.pivot_table(index='Year', columns='Month', aggfunc='size')
```

```python
flightvol_months
```

```python
# Create heatmap
plt.figure(figsize=(8,6))
sns.heatmap(flightvol_months, cmap='viridis_r', linecolor='black', linewidth=1)
plt.xticks(ticks=range(1, 13), labels=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],rotation=45)
plt.title('Flight Volume by Year and Month',fontsize=15,fontstyle='oblique')
plt.show()
```

```python
# Extract day of week and hour from 'FL_DATE'
Flightsdf['Day_of_Week'] = Flightsdf['Fl_date'].dt.dayofweek
Flightsdf['Hour'] = Flightsdf['Crs_dep_time'] // 100  # Extract hour from CRS_DEP_TIME

# Creating pivot table to aggregate flight volume by day of week and hour
flight_volume = Flightsdf.pivot_table(index='Hour', columns='Day_of_Week', aggfunc='size')
```

```python
flight_volume
```

```python
#heatmap of Flight Volume by Day and Hour
plt.figure(figsize=(8,6))
sns.heatmap(flight_volume, cmap='viridis_r', linecolor='black', linewidth=0.5)
plt.xticks(ticks=range(7), labels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],rotation=45)
plt.title('Flight Volume by Day and Hour',fontsize=15,fontstyle='oblique')
plt.show()
```

```python
#Plot the average arrival delay for each day of the week over time.
sns.lineplot(y=Flightsdf['Arr_delay'],x=Flightsdf['Fl_date'].dt.dayofweek,color='#A267AC')
plt.xticks(ticks=range(7), labels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Average Arrival Delay by Day of Week',fontsize=15,fontstyle='oblique')
plt.show()
```
